```python
from datetime import datetime
df['month'] = pd.DatetimeIndex(df['START_DATE']).month

month_label ={1.0:'jan',2.0:'feb',3.0:'mar',4.0:'april',5.0:'may',6.0:'june',7.0:'july',8.0:'august',9.0:'sep',10.0:'oct',11.0:'nov',12.0:'dec'}
df["month"] = df.month.map(month_label)
mon = df.month.value_counts(sort = False)
```

```python
df.head()
```

```python
fd = pd.DataFrame({
 "monthss" : mon.values,
  "value count" : df.groupby('month',sort =  False)['MILES'].max()
    
})

p = sns.lineplot(data = fd)
p.set(xlabel = "monthss",ylabel = "value count")
plt.show()
```

# 5)on which day of the week people book ubers ride the most

```python
df['day'] = df.START_DATE.dt.weekday
data_label ={

    0:'mon',1:'tue',2:'web',3:'thu',4:'fri',5:'sat',6:'sun'
    
}

df['day'] = df['day'].map(data_label)
```

```python
df.head()
```

```python
day_label = df.day.value_counts()

sns.barplot(x = day_label.index,y =day_label)
plt.xlabel =['day']
plt.ylabel =['count']
```

#### Peak demand occurs on Fridays, suggesting:
#### End-of-week travel for work, errands, or social plans
#### Possibly evening events, hangouts, or commutes
#### Weekdays have consistent usage patterns â€” possibly commuters or business travel
#### Weekend dips imply less structured movement or users relying on other modes of transport
# recommdations
#### Increase driver availability on Fridays, especially in the evening hours.
#### Offer Friday-specific promotions or surge pricing to capture the peak demand.
